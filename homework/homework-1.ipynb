```python
end = date.today()
print(f'Year = {end.year}; month= {end.month}; day={end.day}')

start = date(year=end.year-70, month=end.month, day=end.day)
print(f'Period for indexes: {start} to {end} ')
```

```python
# Real Potential Gross Domestic Product (GDPPOT), Billions of Chained 2012 Dollars, QUARTERLY
# https://fred.stlouisfed.org/series/GDPPOT
#gdppot = pdr.DataReader("GDPPOT", "fred", start=start)

# Real Gross Domestic Product (GDPC1), Billions of Chained 2017 Dollars, QUARTERLY
# https://fred.stlouisfed.org/series/GDPC1
gdpc1 = pdr.DataReader("GDPC1", "fred", start=start)
```

```python
gdpc1['gdpc1_us_yoy'] = gdpc1.GDPC1/gdpc1.GDPC1.shift(4)-1
gdpc1['gdpc1_us_qoq'] = gdpc1.GDPC1/gdpc1.GDPC1.shift(4)-1
gdpc1last4 = gdpc1.tail(4)
avg_growth_gdp_2023 = gdpc1last4['gdpc1_us_yoy'].mean()
print(avg_growth_gdp_2023)
```

```python
#Question 4
q4_dct = {'2222.SR':None, 'BRK-B':None, 'AAPL':None, 'MSFT':None, 'GOOG':None, 'JPM':None}

for x in q4_dct.keys():
  q4_daily = yf.download(tickers = x,
                      start="2023-01-01",
                      end="2023-12-31",
                      interval = "1d")
  max_adj_close_q4_daily = q4_daily['Adj Close'].max()
  min_adj_close_q4_daily = q4_daily['Adj Close'].min()
  lrr_adj_close_q4_daily = (max_adj_close_q4_daily-min_adj_close_q4_daily)/max_adj_close_q4_daily
  q4_dct[x] = lrr_adj_close_q4_daily

q4_max_key = max(q4_dct, key=q4_dct.get)
print("\n", q4_max_key, "has the largest range ratio at", round(q4_dct.get(q4_max_key), 2))
```

[Question 4] MSFT 0.42

```python
#refactor - multiple tickers in one set
q4_2_dct = {'2222.SR':None, 'BRK-B':None, 'AAPL':None, 'MSFT':None, 'GOOG':None, 'JPM':None}
q4_2_daily = yf.download(tickers = list(q4_2_dct.keys()),
                      start="2023-01-01",
                      end="2023-12-31",
                      interval = "1d")
#q4_2_daily.tail()

for x in q4_2_dct.keys():
  max_adj_close_q4_2_daily = q4_2_daily['Adj Close'][x].max()
  min_adj_close_q4_2_daily = q4_2_daily['Adj Close'][x].min()
  lrr_adj_close_q4_2_daily = (max_adj_close_q4_2_daily-min_adj_close_q4_2_daily)/max_adj_close_q4_2_daily
  q4_2_dct[x] = lrr_adj_close_q4_2_daily

q4_2_max_key = max(q4_2_dct, key=q4_2_dct.get)
print("\n", q4_2_max_key, "has the largest range ratio at", round(q4_2_dct.get(q4_2_max_key), 2))
```

```python
#Question 5
q5_dct = {'2222.SR':None, 'BRK-B':None, 'AAPL':None, 'MSFT':None, 'GOOG':None, 'JPM':None}

for x in q5_dct.keys():
  q5_ticker = yf.Ticker(x)
  q5_dividends_2023 = q5_ticker.dividends[q5_ticker.dividends.index.year == 2023].sum()
  q5_daily = yf.download(tickers = x,
                      start="2023-01-01",
                      end="2023-12-31",
                      interval = "1d")
  q5_adj_close_eoy_2023 = q5_daily.loc[q5_daily.index[-1]]['Adj Close']
  q5_yield_2023 = q5_dividends_2023/q5_adj_close_eoy_2023
  q5_dct[x] = q5_yield_2023

q5_max_key = max(q5_dct, key=q5_dct.get)
print("\n", q5_max_key, "has the largest yield value at", round(q5_dct.get(q5_max_key)*100, 1), "%")
```
